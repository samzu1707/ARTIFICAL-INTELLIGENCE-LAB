### **_DYNAMIC HANGMAN_** ###

```python
import random

Hangman_Stages = [
   """
     ------
     |    |
     |    
     |   
     |    
     |
    ===
    """,
    """
     ------
     |    |
     |    O
     |   
     |    
     |
    ===
    """,
    """
     ------
     |    |
     |    O
     |    |
     |    
     |
    ===
    """,
    """
     ------
     |    |
     |    O
     |   /|
     |    
     |
    ===
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |    
     |
    ===
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / 
     |
    ===
    """,
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |
    ===
    """
]

Words= [
    ("grapes","Fruit"),
    ("cherry","Fruit"),
    ("banana","Fruit"),
    ("buffalo","Animal"),
    ("tiger","Animal"),
    ("elephant","Animal"),
    ("delhi","City"),
    ("karachi","City"),
    ("milano","City"),
    ("england","Country"),
    ("pakistan","Country"),
    ("spain","Country")
]

Word, Hint = random.choice(Words)
Guessed = ["_"] * len(Word)

Num_of_Attempts = len(Hangman_Stages) - 1

print("Welcome to Hangman!")
print(f"Hint:The word is a {Hint}")

while Num_of_Attempts > 0 and "_" in Guessed:
    print(Hangman_Stages[len(Hangman_Stages) - 1 - Num_of_Attempts])
    print("Word: "," ".join(Guessed))
    print(f"Number of Attempts left: {Num_of_Attempts}")

    guess = input("Guess a letter: ").lower()

    if guess in Word:
        for i, letter in enumerate(Word):
            if letter == guess:
                Guessed[i] = guess
        print("Good guess")
    else:
        Num_of_Attempts -= 1
        print("Wrong guess")

if "_" not in Guessed:
    print("\n You WON! The word was:",Word)
else:
    print(Hangman_Stages[0])
    print("\n You LOST! The word was:",Word)
```
